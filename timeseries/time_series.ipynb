```python
import pandas as pd
import numpy as np
date = pd.to_datetime("4th of July, 2015")
date
```

```python
# day of the week
date.strftime('%A')
```

```python
# adding and creating a DatetimeIndex
new_ = date + pd.to_timedelta(np.arange(12), 'D')
type(new_)
new_
```

```python
index = pd.DatetimeIndex(['2014-07-04', '2014-08-04',
                          '2015-07-04', '2015-08-04'])
data = pd.Series([0, 1, 2, 3], index=index)

data
```

```python
# creates a series of 365 days
starting_date = pd.to_datetime("1st of January, 2021")
# creating a datetimeindex
new_datetimeindex = starting_date + pd.to_timedelta(np.arange(365), 'D')
new_index = pd.DatetimeIndex(new_datetimeindex)

new_dataseries = pd.Series(range(0, 365), index=new_index)


new_dataseries
```

```python
new_dataseries['2021-04-20':'2021-05-10']
```

```python
from pandas_datareader import data
data.DataReader

snowflake = data.DataReader('SNOW', start='2019-05-10', end='2021-05-15',
                       data_source='yahoo')
snowflake.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
snowflake.plot()
```

```python
snowflake.plot(alpha=0.5, style='-')
snowflake.resample('BA').mean().plot(style=':')
snowflake.asfreq('BA').plot(style='--');
# snowflake.legend(['input', 'resample', 'asfreq'],
#            loc='upper left');
```

```python
fig, ax = plt.subplots(2, sharex=True)
data = snowflake.iloc[:10]

data.asfreq('D').plot(ax=ax[0], marker='o')

data.asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
ax[1].legend(["back-fill", "forward-fill"]);
```

```python
fig, ax = plt.subplots(3, sharey=True)

# apply a frequency to the data
snowflake = snowflake.asfreq('D', method='pad')

snowflake.plot(ax=ax[0])
snowflake.shift(900).plot(ax=ax[1])
snowflake.shift(900).plot(ax=ax[2])

# legends and annotations
local_max = pd.to_datetime('2007-11-05')
offset = pd.Timedelta(900, 'D')

ax[0].legend(['input'], loc=2)
ax[0].get_xticklabels()[2].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=0.3, color='red')

ax[1].legend(['shift(900)'], loc=2)
ax[1].get_xticklabels()[2].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')

ax[2].legend(['shift(900)'], loc=2)
ax[2].get_xticklabels()[1].set(weight='heavy', color='red')
ax[2].axvline(local_max + offset, alpha=0.3, color='red');
```
