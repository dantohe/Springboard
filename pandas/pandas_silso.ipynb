```python
import boto3
import pandas as pd
import io
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
DATASET = '/Users/dantohe/Development/Projects/Springboard/datasets/silso/SN_d_tot_V2.0.csv'
sunspots = pd.read_csv(DATASET, sep = ';', header = None)
sunspots.info()
```

```python
sunspots.head()
sunspots.iloc[10:15, :]
```

```python
columns = ['Year', 'Month', 'Day', 'fr_of_year','Daily_total_sunspot_number', 'Daily_standard_deviation', 'Number_of_observations_used_daily','Definitive']
print(len(columns))
# amalgamating the first columns into a date
# parse_dates=[[0,1,2]]
# dealing with NA values
# na_values = [' -1','  -1'],

sunspots = pd.read_csv(DATASET, sep = ';', header = None, names = columns, na_values = [' -1','  -1'], parse_dates=[[0,1,2]])
sunspots.info()
```

```python
sunspots.iloc[10:15]
```

```python
sunspots.index = sunspots['Year_Month_Day']
sunspots.index.name = 'date'
sunspots.info()
```

```python
# trimming
cols = ['Daily_total_sunspot_number','Definitive']
sunspots = sunspots[cols]
sunspots.info()
sunspots.iloc[10:20,:]
```

```python
# export
export_path = '/Users/dantohe/Development/Projects/Springboard/datasets/silso/export'
csv_fname = 'exported_sunspots.csv'
xlsx_name = 'exported_sunspots.xlsx'

if not os.path.exists(export_path):
    os.mkdir(export_path)

csv_export_path = os.path.join(export_path, csv_fname)    
xlsx_export_path = os.path.join(export_path, xlsx_name)    

sunspots.to_csv(csv_export_path)
# sunspots.to_excel(xlsx_export_path)
```
